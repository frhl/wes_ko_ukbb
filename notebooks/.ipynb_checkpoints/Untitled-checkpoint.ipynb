```r
library(data.table)
library(ggplot2)
library(Hmisc)
```

```r
files <- sort(list.files('data/switch', pattern = '.txt', full.names = TRUE))
files <- files[grepl('chr21',files)]
files
```

```r
d <- fread('data/switch/ukb_wes_200k_call_scaffold_chr20.txt')
```

```r
aggregate_ser_by_position <- function(d){
    
    d1 <- aggregate(switches ~ POS, data = d, FUN = sum)
    d2 <- aggregate(switch ~ POS, data = d, FUN = length)
    d3 <- d[,c('POS','MAF','HWE')]
    d3 <- d3[!duplicated(d3),]
    
    colnames(d1) <- c('POS','n_switch')
    colnames(d2) <- c('POS','n_tested')
    mrg <- merge(merge(d1, d2), d3)
    mrg$n_switch_div_n_tested <- mrg$n_switch/mrg$n_tested
    return(mrg)
    
}

```

```r
Ds <- lapply(files, function(f){
    d <- fread(f)
    if ('MAF' %in% colnames(d)){
        d <- aggregate_ser_by_position(d)
        d$name <- basename(f)
        return(d)
    } 
})
```

```r
d <- do.call(rbind, Ds)
```

```r
# extract switch errors
d_summary <- cbind(
    data.frame(aggregate(n_tested ~ name, data = d, FUN = sum)),
    data.frame(n_switch = aggregate(n_switch ~ name, data = d, FUN = sum)$n_switch)
    )
```

```r
bconfs <- do.call(rbind, lapply(1:nrow(d_summary), function(i){
    
    M <- d_summary[i,]
    binconf(M$n_switch, M$n_tested)
     
}))

d_summary <- cbind(
    d_summary,
    bconfs
)
```

```r
keep <- d_summary$name[!duplicated(d_summary[,c('n_tested','n_switch')])]
d_summary <- d_summary[d_summary$name %in% keep,]
```

```r
options(repr.plot.width=12, repr.plot.height=5)
ggplot(d_summary, aes(y = reorder(name, PointEst), x = PointEst, xmax=Upper, xmin=Lower)) +
  geom_bar(stat="identity", color="black", fill = 'yellow', position=position_dodge()) +
  geom_errorbar(width=.2, position=position_dodge(.9)) +
  xlab('Switch Errors Point Estimate') +
  ylab('Phasing strategy') +
  ggtitle('Switch errors across phasing strategies')
```

```r
p2 <- ggplot(d_summary, aes(y = log10(n_tested), x = PointEst, xmax=Upper, xmin=Lower, color = name)) +
  geom_point(size = 2) + 
  geom_errorbar(width=.01, position=position_dodge(.9), color = 'black') + 
  xlab('Switch Errors Point Estimate') +
  ylab('log10(Sites tested)') +
  ggtitle('Switch errors versus sites tested')

print(p2)
p2b <- p2 + xlim(0,0.01)
print(p2b)
```
