## Comparison of phased and unphased WES data

```python
# load data
path1 = '/well/lindgren/UKBIOBANK/flassen/projects/KO/wes_ko_ukbb/data/qc/ukb_wes_200k_chr21_variants_unphased.ht'
path2 = '/well/lindgren/UKBIOBANK/flassen/projects/KO/wes_ko_ukbb/data/qc/ukb_wes_200k_chr21_variants_phased.ht'
mt1 = hl.read_table(path1)
mt2 = hl.read_table(path2)
```

```python
# load VEP annotation table
input_annotation_path = "data/vep/hail/ukb_wes_200k_chr21_vep.ht"
consequence_annotations = hl.read_table(input_annotation_path)
mt1 = mt1.annotate(consequence = consequence_annotations[mt1.key]) 
mt2 = mt2.annotate(consequence = consequence_annotations[mt2.key]) 
```

```python
# annotate consequence category
annotation_by_variant1 = analysis.annotation_case_builder(mt1.consequence.vep.worst_csq_for_variant_canonical, mt1.consequence.dbnsfp)
annotation_by_variant2 = analysis.annotation_case_builder(mt2.consequence.vep.worst_csq_for_variant_canonical, mt2.consequence.dbnsfp)
mt2 = mt2.annotate(consequence_category = annotation_by_variant2)
mt1 = mt1.annotate(consequence_category = annotation_by_variant1)
```

```python
# subset by PTV
mt2 = mt2.filter(mt2.consequence_category == 'ptv')
mt1 = mt1.filter(mt1.consequence_category == 'ptv')
```

```python
# How many samples are at homozygous or heterozygous PTVs?
print(mt1.count())
print(mt2.count())
```

```python
# How many heterozygous calls
hm1 = mt1.variant_qc.homozygote_count[1].collect()
hm2 = mt2.variant_qc.homozygote_count[1].collect()
```

```python
#mt1.describe()
hm1 = mt1.variant_qc.homozygote_count[1].collect()
hm2 = mt2.variant_qc.homozygote_count[1].collect()
```

```python
print(sum(hm1))
print(sum(hm2))
```

```python
# Looks like there is only three sites where samples are homozygous
print(sum([int(x)!=0 for x in hm1]))
print(sum([int(x)!=0 for x in hm2]))
```

```python
# show these three sites in each dataset
mt1.filter(mt1.variant_qc.homozygote_count[1] > 0).show()
mt2.filter(mt2.variant_qc.homozygote_count[1] > 0).show()
```

*conclusion:* Three sites PTV sites have homozygous carrieres in our data.

## Just some trashy helper functions i had to set up to avoid using dbNSFP (which are only used for missense anyway)

```python
#### HELPER FUNCTIONS ####

PLOF_CSQS = ["transcript_ablation", "splice_acceptor_variant",
             "splice_donor_variant", "stop_gained", "frameshift_variant"]

MISSENSE_CSQS = ["stop_lost", "start_lost", "transcript_amplification",
                 "inframe_insertion", "inframe_deletion", "missense_variant"]

SYNONYMOUS_CSQS = ["stop_retained_variant", "synonymous_variant"]

OTHER_CSQS = ["mature_miRNA_variant", "5_prime_UTR_variant",
              "3_prime_UTR_variant", "non_coding_transcript_exon_variant", "intron_variant",
              "NMD_transcript_variant", "non_coding_transcript_variant", "upstream_gene_variant",
              "downstream_gene_variant", "TFBS_ablation", "TFBS_amplification", "TF_binding_site_variant",
              "regulatory_region_ablation", "regulatory_region_amplification", "feature_elongation",
              "regulatory_region_variant", "feature_truncation", "intergenic_variant"]

def annotation_case_builder_old(worst_csq_by_gene_canonical_expr, use_loftee: bool = True):
    case = hl.case(missing_false=True)
    if use_loftee:
        case = (case
             .when(worst_csq_by_gene_canonical_expr.lof == 'HC', 'ptv')
             .when(worst_csq_by_gene_canonical_expr.lof == 'LC', 'ptv_LC')
            )
    else:
        case = case.when(hl.set(PLOF_CSQS).contains(worst_csq_by_gene_canonical_expr.most_severe_consequence), "ptv")
    case = (case.when(hl.set(MISSENSE_CSQS).contains(worst_csq_by_gene_canonical_expr.most_severe_consequence), "missense")                                   
                .when(hl.set(MISSENSE_CSQS).contains(worst_csq_by_gene_canonical_expr.most_severe_consequence), "other_missense")
                .when(hl.set(SYNONYMOUS_CSQS).contains(worst_csq_by_gene_canonical_expr.most_severe_consequence), "synonymous")
                .when(hl.set(OTHER_CSQS).contains(worst_csq_by_gene_canonical_expr.most_severe_consequence), "non_coding")
           )
    return case.or_missing()
```
