```python
#hl.import_bed('data/file1.bed', reference_genome='GRCh37')
AUTOSOMES = list(map(str, range(1, 23)))
```

```python
def get_ukb_genotypes_bed_path(chrom):
    return f'/well/lindgren/UKBIOBANK/DATA/CALLS/ukb_cal_chr{chrom}_v2.bed'

def get_ukb_genotypes_bim_path(chrom):
    return f'/well/lindgren/UKBIOBANK/DATA/CALLS/ukb_snp_chr{chrom}_v2.bim'

def get_ukb_genotypes_fam_path(chrom):
    return "/well/lindgren/UKBIOBANK/DATA/SAMPLE_FAM/ukb11867_cal_chr1_v2_s488363.fam"

def get_ukb_genotypes_bed(chroms: list = AUTOSOMES):
    r"""Import UK Biobank genotype calls as a MatrixTable from plink files
    
    :param chroms: List of chromosomes to import
    """
    all_datasets = list()
    for chrom in chroms:
        bed_path = get_ukb_genotypes_bed_path(chrom)
        bim_path = get_ukb_genotypes_bim_path(chrom)
        fam_path = get_ukb_genotypes_fam_path(chrom)
        bed = hl.import_plink(
            bed = bed_path, 
            bim = bim_path, fam = fam_path, 
            reference_genome='GRCh37', 
            contig_recoding={str(chrom).zfill(2): str(chrom)}
        )
        all_datasets.append(bed)
    dataset_result = hl.MatrixTable.union_rows(*all_datasets)
    return dataset_result
```
