```r
args <- list(
    dir = "data/prs/scores/auto",
    phenotype = "BC_combined"
)
```

```r
files <- list.files(args$dir, pattern = paste0(args$phenotype,"_chr[0-9]+.txt.gz"), full.names = TRUE)
n <- length(files)

```

```r
d <- do.call(rbind, lapply(1:22, function(x) {
      f <- files[grepl(paste0("chr",x,".txt.gz"),files)]
      if (length(f)){
          d <- fread(f)
          d$chr <- paste0("chr",x)
          return(d)
      } else {
        write(paste0("Warning: ", pheno, "chr", x," does not exists!"), stderr())
      }
  }))
```

```r
# Aggregate PRS into a n x 22 matrix (n = samples)
M <- dcast(sid ~ chr, value.var = "prs", data = d)
M <- M[,c('sid', paste0("chr",1:22))]
```

```r
# sum up PRS score
M[is.na(M)] <- 0
```

```r
M <- data.table(
  sid = M$sid,
  pgs = rowSums(M[,2:ncol(M)]))
colnames(M)[-1] <- paste0(args$phenotype, '_', colnames(M)[-1])
```
