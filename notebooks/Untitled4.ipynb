```r
samples <- fread("data/phenotypes/filtered_phenotypes_cts.tsv.gz")
```

```r
samples <- samples$eid[samples$genetic.eur.no.fin.oct2021]
```

```r
fs <- list.files('data/knockouts/alt/', pattern = "pLoF_damaging_missense.tsv.gz", full.names = TRUE)
```

```r
fread(fs[1])
```

```r
read_ko <- function(f){
    d <- fread(f)
    d <- d[,c(2,1,9,10), with = FALSE]
    d <- d[d$s %in% samples]
    in_genes <- unique(d$gene_id)
    in_samples <- unique(d$s)
    out_samples <- samples[!samples %in% in_samples]
    grid <- expand.grid(out_samples, in_genes)
    colnames(grid) <- c("s","gene_id")
    grid$pKO <- 0
    grid$knockout <- NA
    mrg <- rbind(d, grid)
    return(mrg)
}
```

```r
for (f in fs){
    d <- read_ko(f)
    chrom <- stringr::str_extract(f, "chr[0-9]+")
    print(chrom)
    genes <- unique(d$gene_id)
    thedir <- paste0("data/survival/knockouts/eur_no_fin/pLoF_damaging_missense/",chrom)
    dir.create(thedir)
    for (g in genes){
        d_gene <- d[d$gene_id %in% g,]
        if (sum(d_gene$pKO) >= 4){
            outfile = paste0(thedir,"/",g,".txt.gz")
            fwrite(d_gene, outfile, sep = '\t', quote = FALSE)
        }
    }
}
```
