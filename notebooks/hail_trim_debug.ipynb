```python
def sort_by_chunks(chunks):
    ''' sort by X in XofY with regex '''
    n = len(chunks)
    lst = { int(re.findall(r'\d+of\d+', c)[0].split("of")[0])-1 : c for c in chunks }
    chunks = [lst[i] for i in range(n)]
    return chunks
```
