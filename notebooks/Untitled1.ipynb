```r
d <- fread('data/permute/genes/chr21/ukb_eur_wes_200k_pLoF_damaging_missense_chr21_ENSG00000141956.tsv.gz')
```

```r
# simple method but still a bit slow
shuffle_knockouts <- function(d){
    d$KO <- rbinom(n=nrow(d), size=1, prob = d$pTKO)
    d$pKO <- ifelse((d$KO == 1), 1,
             ifelse((d$phased_het == 1 & d$unphased_het > 0), 1 - 1*(1/2)^d$unphased_het,
             ifelse((d$phased_het == 0 & d$unphased_het > 1), 1 - 2*(1/2)^d$unphased_het, 0)))
    return(d$pKO)
}
```

```r

detectCores()
cl <- makeCluster(detectCores()-1)  
clusterEvalQ(cl,library(data.table))
clusterExport(cl,c("d"))
clusterExport(cl,c("shuffle_knockouts"))
```

```r
system.time(parLapply(cl, 1:48, function(i) { 
    set.seed(i)
    samp <- replicate(10, shuffle_knockouts(d))
    rownames(samp) <- d$s
    samp <- data.table(t(samp))
    return(samp)
} ))
```
